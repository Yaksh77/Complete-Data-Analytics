```python
# Airthmentic Operators

a = 10
b = 15

add_result = a + b
sub_result = a - b
mul_result = a * b
div_result = (a / b)
# div_result = (a / b).__round__(2)
floor_div_result = a // b # floor division
moduls_result = a % b
expon_result = a ** b

print(add_result, sub_result, mul_result, div_result, floor_div_result, moduls_result, expon_result)
```
