```python
age = 20
height = 5.8
name = "John"
is_student = True

print("Age :",age)
print("Height :",height)
print("Name :",name)
print(is_student)
```

```python
## Naming Conventions
## snake_case
## Variables names are case sensitive

# Invalid variable names
# 1kg = 5
# first-name = "Yaksh"
# @name = "Yaksh"
# print(@name)
# print(first-name)
# print(1kg)

first_name = "Yaksh"
last_name = "Mahadev"
print(first_name)
print(last_name)

```

```python
## Understanding variable types

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
print(type(first_name))
print(type(last_name))
```

```python
# Type Checking and Type Conversion

age = 20
print(type(age))

age = str(age)
print(type(age))
print(age)

age = "20"
print(type(age))

age = int(age)
print(type(age))
print(age)
```

```python
# Dynamic Typing
# Python allows the type of a variable to change as the program executes
var = 10
print(var,type(var))

var = "Hello"
print(var,type(var))

var = True
print(var,type(var))

```

```python
# Input taking from user

age = int(input("Enter your age : "))
print(age)
print(type(age))
```
