```python
## Basic Syntax rules in python
## Case sensitivity - python is case sensitive

name = "Yaksh"
Name = "Mahadev"

print(name)
print(Name)

```

```python
### Indetation

age = 20
if age >= 18:
    print("You can drive")
else:
    print("You can't drive")
print(age)

```

```python
## This is a single line comment
print("Har Har Mahadev")
```

```python
## Line Continutation
## Use a backslash(\) to continue a line

total = 1 + 2 + 3 + \
        4 + 5 + 6
print(total)
```

```python
## Multiple statements in single line

x = 5; y = 6; z = x + y
print(z)
```

```python
# Variable Assignment

x, y, z = 1, 2, 3
print(x)
print(y)
print(z)

age = 55
name = "Ramesh"
print(name)
```
