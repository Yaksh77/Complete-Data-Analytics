```python
# syntax
# lamda arguments or parameters: expression
```

```python
addition = lambda a,b: a + b
addition(1,2)
```

```python
def even(num):
    if num % 2 == 0:
        return True
    else:
        return False
even(2)
```

```python
even1 = lambda num: num % 2 == 0
even1(2)
```

```python
# map function
list(map(lambda x: x ** 2, [1,2,3,4,5]))
```
