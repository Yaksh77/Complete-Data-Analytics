```python
to_do_list = ["Buy milk", "Play video games", "Go to gym"]

# Adding to task
to_do_list.append("Go for a run")

# Removing task
to_do_list.remove("Play video games")

if "Go to gym" in to_do_list:
    print("Go to gym is in the list")
else:
    print("Go to gym is not in the list")

for task in to_do_list:
    print(f"Tasks : {task}")
```
