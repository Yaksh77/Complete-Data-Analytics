```python
def even_or_odd(number):
    """
    Check if a number is even or odd
    """
    if number % 2 == 0:
        return "Given number is even"
    else:
        return "Given number is odd"
even_or_odd(3)
```

```python
# function with multiple parameters
def add(a, b):
    """
    Add two numbers
    """
    return a + b

add(1, 2)
```

```python
# Default parameters
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}")

greet("Yash")
```

```python
# Variable length arguments
# Postional arguments and keywords arguments

def print_numbers(*args):
    print(args) # Gives args tuple
    # print(args[0])
    for number in args:
        print(number)

print_numbers(1, 2, 3,"Yaksh")
```

```python
# keywords arguments

def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(name="Yash", age=20,country="India")
```

```python
# Combining both arguments
def print_details(*args, **kwargs):
    print(args)
    print(kwargs)

print_details(1, 2, 3, name="Yash", age=20, country="India")
```

```python
# Return statements
def multiply(a, b):
    return a * b

result = multiply(2, 3)
print(result)
```
