```python
# for loop
for i in range(5):
    print(i);
```

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(10,1,-2):
    print(i)
```

```python
str = "Yaksh Hello"
for i in str:
    print(i)
```

```python
# While loop
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
# Loop control statements

# break
for i in range(10):
    if(i == 5):
        break;
    print(i)
```

```python
# continue statement

for i in range(10):
    if(i == 5):
        continue
    print(i)
```

```python
# pass statement


for i in range(10):
    if i == 3:
        pass
    print(i)
```

```python
# Nested Loops

for i in range(5):
    for j in range(5):
        for k in range(5):
            print(i,j,k)
```

```python
for num in range(1,51):
    if num > 1:
        for i in range(2,num):
            if num % i == 0:
                break;
        else:
            print(num)
```
