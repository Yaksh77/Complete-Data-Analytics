```python
# Accessing elements of dictionary
dictionary = {'Name': 'Zara', 'Age': 7}
print("dictionary['Name']: ", dictionary['Name'])
print("dictionary['Age']: ", dictionary['Age'])

# Using get() method
print(dictionary.get('Name'))
print(dictionary.get('grade'))
print(dictionary.get('grade',"Not Found"))
```

```python
# Modifying dictionary elements 
# Dictionary is mutable

# Adding element to the dictionary
dictionary['Grade'] = 8
print(dictionary)

# Removing element from the dictionary
dictionary.pop('Grade')
print(dictionary)
```

```python
# Another way to remove element
del dictionary['Name']
print(dictionary)
```

```python
# Dcitionary Methods

# clear()	Removes all the elements from the dictionary
# copy()	Returns a copy of the dictionary
# fromkeys()	Returns a dictionary with the specified keys and value
# get()	Returns the value of the specified key
# items()	Returns a list containing a tuple for each key value pair
# keys()	Returns a list containing the dictionary's keys
# pop()	Removes the element with the specified key
# popitem()	Removes the last inserted key-value pair
# setdefault()	Returns the value of the specified key. If the key does not exist: insert the key, with the specified value
# update()	Updates the dictionary with the specified key-value pairs
# values()	Returns a list of all the values in the dictionary

keys = dictionary.keys()
values = dictionary.values()
print(keys)
print(values)

items = dictionary.items()
print(items)
```

```python
# Shallow Copy
dict_copy = dictionary
print(dict_copy)
print(dictionary)
```

```python
dictionary['name'] = "Yaksh"
print(dict_copy)
print(dictionary)
```

```python
# Shallow Copy
dict_copy1 = dictionary.copy()
dict_copy1['name'] = "Yash"
print(dict_copy1)
print(dictionary)
```

```python
# Iterating over dictionaries
for key in dictionary:
    print(key, dictionary[key])
    
for key in dictionary.keys():
    print(key)
    
for value in dictionary.values():
    print(value)
    
for key, value in dictionary.items():
    print(key, value)
```

```python
# Nested Dictionaries
students = {
    "student1":{
        "name": "Yaksh",
        "marks": 90
    },
    "student2":{
        "name": "Yash",
        "marks": 70
    },
    "student3":{
        "name": "Harsh",
        "marks": 80
    }
}
print(students)
```

```python
# Access Nested dictionaries
print(students["student1"]["marks"])
```

```python
# Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(student_id, student_info)
    for key, value in student_info.items():
        print(key, value)
```

```python
# Dictionary Comprehension
d = {x: x**2 for x in range(10)}
print(d)
```

```python
d = {x: x**2 for x in range(10) if x % 2 == 0}
print(d)
```

```python
numbers = [1,2,2,3,3,4,4,4,5,5]
# frequency = {x:numbers.count(x) for x in numbers}
frequency = {}

for number in numbers:
    if (number in frequency):
        frequency[number] += 1;
    else:
        frequency[number] = 1;
print(frequency)
```
