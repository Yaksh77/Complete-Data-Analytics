```python
# Creating a tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(lst)
print(type(lst))
tup = tuple()
print(tup)
print(type(tup))
```

```python
numbers = tuple([1, 2, 3, 4, 5])
print(numbers)
print(type(numbers))
```

```python
list(numbers)
```

```python
mixed_tupe = (1, 2, 3, 4, 5, "a", "b", "c", "d", "e")
print(mixed_tupe)
```

```python
# Accessing Tuple elements
print(numbers)
print(numbers[0])
```

```python
# Tuple operations
concate = numbers + mixed_tupe
print(concate)
```

```python
mixed_tupe * 3
```

```python
# Immuable nature
mixed_tupe[0] = "x"
print(mixed_tupe)
```

```python
# Tuple methods
print(numbers)
print(numbers.count(1))
print(numbers.index(5))
```

```python
# Packing and Unpacking
# Packing
packed_tuple = 1, "Hello", 3.14
print(packed_tuple)

# Unpacking
a, b, c = packed_tuple
print(a)
print(b)
print(c)

# unpacking with *
numbers = (1,2,3,4,5,6)
a, b, *c = numbers
print(a)
print(b)
print(c)
```

```python
lst = [[5],[],[],(1,2,3)]

# Nested Tuple
nested_tuple = ((1,2,3),(4,5,6),(7,8,9))
print(nested_tuple)
print(nested_tuple[0][1])
```

```python
# Iterating over Tuple
for i in nested_tuple:
    for j in i:
        print(j,end=" ")
    print()
```
